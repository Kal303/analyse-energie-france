```python
from google.colab import files
uploaded = files.upload()
import pandas as pd

df = pd.read_csv("energie.csv")

print(df.head())

from scipy.stats import ttest_ind

#On filtre les données pour le gaz uniquement
gaz = df[df['filiere'] == 'Gaz']

#On sélectionne les deux régions
idf = gaz[gaz['nom_region'] == 'Île-de-France']['conso_totale_mwh']
hdf = gaz[gaz['nom_region'] == 'Hauts-de-France']['conso_totale_mwh']

#T-test
t_stat, p_value = ttest_ind(idf, hdf)
print("Statistique t :", t_stat)
print("P-value :", p_value)

#Régression

import statsmodels.api as sm

idf_gaz = gaz[gaz['nom_region'] == 'Île-de-France']
X = idf_gaz['annee']
y = idf_gaz['conso_totale_mwh']
X = sm.add_constant(X) # Ajoute l'intercept

model = sm.OLS(y, X).fit()
print(model.summary())
```
